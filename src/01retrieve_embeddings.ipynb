```python
import pandas as pd
import openai
from dotenv import load_dotenv
import tqdm
```

```python
# load environment variables from a .env file (openai api key, etc.)
load_dotenv()
```

```python
df = pd.read_excel("../data/openalex_systematic_review.xlsx")
```

```python
client = openai.OpenAI()
```

```python
embeddings = []

# generates embeddings for each abstract in the dataframe
for abstract in tqdm.tqdm(df["abstract"].tolist()):
    response = client.embeddings.create(
        model="text-embedding-3-large",
        input=abstract
    )
    embeddings.append(response.data[0].embedding)
```

```python
# adds the embeddings as a new column in the dataframe
df["embedding"] = embeddings
```

```python
df_embeddings = df[["id", "doi", "title", "abstract", "primary_location.source.display_name", "embedding"]]
```

```python
df_embeddings.to_csv("../data/papers_with_embeddings.csv", index=False)
```
